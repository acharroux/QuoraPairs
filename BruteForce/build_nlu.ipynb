# use this to load the proper model model: 6 or 50  kinds of questions

```python
import nlu
nlu_questions_50 = nlu.load('en.classify.questions')
nlu_questions_6 = nlu.load('classify.trec6.use')
```

```python
import tqdm
from tqdm import tqdm_pandas
from tqdm.autonotebook  import tqdm
import gc
gc.disable()
gc.collect()


BLOCKSIZE=100000

def fake(df,num_block,first_row,block_size,tag):
    print(num_block,first_row,block_size,df.shape)
    return df

# model with 6 questions
def rename_question_kinds(df,model,tag):
    if model==6:
        df = df.drop(['sentence_embeddings','document','trec6','trec6_sentence_confidence'],axis=1)
        old_start = 'trec6_'
    else:
        df = df.drop(['questions_sentence_confidence','sentence_embeddings','document','questions'],axis=1)
        old_start =  'questions_'
    new_names= dict()
    for c in df.columns:
        if 'confidence' in c:
          new_names[c]=tag+'_'+c.replace('questions_ ','')
    df.rename(columns=new_names,inplace=True)
    return df

# df must be a dataframe with only one column named 'text'
def find_questions_types(df,num_block,first_row,block_size,model,tag):
    if model==6:
      return rename_question_kinds(nlu_questions_6.predict(df,metadata=True,output_level='document'),model,tag)
    else:
      return rename_question_kinds(nlu_questions_50.predict(df,metadata=True,output_level='document'),model,tag)
            
def by_blocks(df,block_size,func,model,tag):
    nb_blocks = int(len(df)/block_size)
    last_size = len(df) % block_size
    res = pandas.DataFrame()
    print('nb_blocks %d block_size %d last_size %d' % (nb_blocks,block_size,last_size))
    assert (nb_blocks*block_size+last_size)==len(df)
    current_row = 0
    for i in tqdm(range(nb_blocks)):
        block = df[current_row:(current_row+block_size)]
        tmp = func(block,i,current_row,block_size,model,tag)
        assert(len(tmp)==block_size), 'bad size of result block'
        res = res.append(tmp,ignore_index=False)
        print('size of result block is %d' % len(res))
        current_row += block_size
        gc.collect()
    if last_size >0:
        block = df[current_row:(current_row+last_size)]
        tmp = func(block,nb_blocks,current_row,last_size,model,tag)
        assert(len(tmp)==last_size), 'bad size of last result block'
        res = res.append(tmp,ignore_index=False)
        print('size of result block is %d' % len(res))
        gc.collect()
    return res

def compute_all_questions_question1(dataframe,dataset_name):
    tag = '1_6'
    print('6 questions type')
    res=by_blocks(dataframe,BLOCKSIZE,find_questions_types,6,tag)
    display(res.head())
    display(res.describe())
    res.to_pickle('./'+dataset_name+'questions_question_'+tag+'.pkl')
    res.to_csv('./'+dataset_name+'questions_question_'+tag+'.csv')
    assert len(res) == len(dataframe), "size of result for 6 is not OK"

    tag = '1_50'
    print('50 questions type')
    res=by_blocks(dataframe,BLOCKSIZE,find_questions_types,50,tag)
    display(res.head())
    display(res.describe())
    res.to_pickle('./'+dataset_name+'questions_question_'+tag+'.pkl')
    res.to_csv('./'+dataset_name+'questions_question_'+tag+'.csv')
    assert len(res) == len(dataframe), "size of result for 50 is not OK"

def compute_all_questions_question2(dataframe,dataset_name):
    tag = '2_6'
    print('6 questions type %s for %s ' % (tag,dataset_name))
    res=by_blocks(dataframe,BLOCKSIZE,find_questions_types,6,tag)
    display(res.describe())
    res.to_pickle('./'+dataset_name+'questions_question_'+tag+'.pkl')
    res.to_csv('./'+dataset_name+'questions_question_'+tag+'.csv')
    assert len(res) == len(dataframe), "size of result for 6 is not OK"

    tag = '2_50'
    
    print('50 questions type %s for %s ' % (tag,dataset_name))
    res=by_blocks(dataframe,BLOCKSIZE,find_questions_types,50,tag)
    display(res.describe())
    res.to_pickle('./'+dataset_name+'questions_question_'+tag+'.pkl')
    res.to_csv('./'+dataset_name+'questions_question_'+tag+'.csv')
    assert len(res) == len(dataframe), "size of result for 50 is not OK"
    
UNITARY_TEST=True
# rehearsal
if UNITARY_TEST:
    train_dataframe = pandas.read_csv(CURRENT_LOCAL_PATH+'train.csv.zip',error_bad_lines=True,warn_bad_lines=True)
    small_train=train_dataframe.sample(10,random_state=42)
    small_train.rename(columns={'question1':'text'},inplace=True)
    small_train=small_train.drop(['question2',"id","qid1","qid2","is_duplicate"],axis=1)
    small_train.fillna('?',inplace=True)
    print(small_train.shape)
    compute_all_questions_question1(small_train,'small_train')
    compute_all_questions_question2(small_train,'small_train')
```
